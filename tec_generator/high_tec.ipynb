```python
%autoreload 2
%matplotlib inline
from tec_map_generator import generate_tec_map, get_ionex_file, get_tec_for_lat_lon_idx
import matplotlib.pyplot as plt

def find_map_location_and_tec(year, day_in_year, hour_in_day, lat_of_interest, lon_of_interest):
    ionex_file = get_ionex_file(year, day_in_year)
    LAT = lat_of_interest
    LON = lon_of_interest
    tecs = generate_tec_map(ionex_file, hour_in_day=hour_in_day, lat_of_interest=lat_of_interest, lon_of_interest=lon_of_interest)
    lat_idx, lon_idx = get_tec_for_lat_lon_idx(tecs, lat_of_interest, lon_of_interest)
    print "TEC VALUE AT {}, {}: {}".format(lat_of_interest, lon_of_interest, tecs[lat_idx][lon_idx])
    plt.show()
```
